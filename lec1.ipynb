<strong style="color: #58a6ff;">üìö CONCEPT: Modern Computer System Structure</strong>
<p style="color: #c9d1d9; margin-top:10px;">A modern general-purpose computer system consists of:</p>
<ul style="color: #c9d1d9;">
<li><strong>One or more CPUs</strong> ‚Äî The processing units</li>
<li><strong>Device Controllers</strong> ‚Äî Each manages specific hardware</li>
<li><strong>Common Bus</strong> ‚Äî Connects everything together</li>
<li><strong>Shared Memory</strong> ‚Äî Accessed by all components</li>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922; margin: 10px 0;">
<strong style="color: #d29922;">‚ö†Ô∏è IMPORTANT: CPU Clarification!</strong>
<p style="color: #c9d1d9; margin-top:10px;">Don't get confused by this! Here's the truth:</p>
<ul style="color: #c9d1d9;">
<li><strong>What people CALL "CPU":</strong> The big box beside your desktop (the cabinet)</li>
<li><strong>What CPU ACTUALLY is:</strong> A small chip on the motherboard!</li>
</ul>
<p style="color: #c9d1d9;">That box contains: Motherboard, CD ROM, USB ports, Power Supply, and more. The <strong>actual CPU</strong> is just a small chip embedded in the motherboard that does all computations! üß†</p>

<strong style="color: #58a6ff;">üìö CONCEPT: Concurrent Execution</strong>
<p style="color: #c9d1d9; margin-top:10px;">The CPU and device controllers can <strong>execute concurrently</strong> ‚Äî they run at the SAME TIME!</p>

<p style="color: #c9d1d9;"><strong>Example:</strong> You're watching a video on monitor WHILE typing in Word document. No lag! Why? Because devices work concurrently. That's the power of modern systems! üöÄ</p>

<p style="color: #c9d1d9;"><strong>Problem:</strong> All devices compete for memory cycles (limited memory!)</p>
<p style="color: #c9d1d9;"><strong>Solution:</strong> Memory Controller synchronizes access so everyone gets their fair share!</p>
</div>

<p style="color: #c9d1d9;">Bootstrap loads the Kernel into RAM ‚Üí Kernel comes alive ‚Üí OS takes over ‚Üí You see your desktop! üñ•Ô∏è</p>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff; margin: 10px 0;">
<strong style="color: #58a6ff;">üìö CONCEPT: Interrupt</strong>
<p style="color: #c9d1d9; margin-top:10px;">The occurrence of an <strong>event</strong> is usually signaled by an <strong>Interrupt</strong> from hardware or software.</p>
<p style="color: #c9d1d9;">Think of it exactly like its English meaning: Something <strong>INTERRUPTS</strong> the CPU while it's working!</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #3fb950; margin: 10px 0;">
<h3 style="color: #3fb950; margin-top:0;">üåç Real-Life Analogy: Interrupt = Phone Call!</h3>
<p style="color: #c9d1d9;">Imagine you're studying (CPU doing work), and suddenly your phone rings (Interrupt arrives). You:</p>
<ol style="color: #c9d1d9;">
<li><strong>Stop studying</strong> (CPU stops current work)</li>
<li><strong>Answer the call</strong> (Execute interrupt task)</li>
<li><strong>Resume studying</strong> after call ends (Resume original work)</li>
</ol>
<p style="color: #c9d1d9;">That's EXACTLY how CPU handles interrupts! üìû</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; margin: 10px 0;">

### üîÑ Types of Interrupts

| Source | Name | How It's Triggered |
|:-------|:-----|:-------------------|
| **Hardware** | **Interrupt** | Hardware sends signal to CPU via System Bus |
| **Software** | **System Call** (Monitor Call) | Software executes a special operation |

</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff; margin: 10px 0;">
<strong style="color: #58a6ff;">üìö CONCEPT: System Call (Monitor Call)</strong>
<p style="color: #c9d1d9; margin-top:10px;">When <strong>software</strong> triggers an interrupt by executing a special operation, it's called a <strong>System Call</strong> or <strong>Monitor Call</strong>.</p>
<ul style="color: #c9d1d9;">
<li><strong>Hardware Interrupt</strong> ‚Üí Just called "Interrupt"</li>
<li><strong>Software Interrupt</strong> ‚Üí Called "System Call" or "Monitor Call"</li>
</ul>
</div>

<h3 style="color: #a371f7; margin-top:0;">üß† Memory Tricks & Mnemonics</h3>

**1. OS Functions: "IRMS"**
- **I**nterface (between user & hardware)
- **R**esource Allocation
- **M**emory Management
- **S**ecurity

**2. OS Goals: "CE" = "Convenience + Efficiency"**
- Think of a "CE" mark on products ‚Äî it means QUALITY!

**3. Bootstrap = "BOOT-STRAP"**
- Computer pulls itself up from sleep by its bootstraps!
- First program ‚Üí Finds OS ‚Üí Loads Kernel ‚Üí System alive!

**4. Interrupt Handling: "STOP ‚Üí JUMP ‚Üí RUN ‚Üí RESUME"**
- **S**top current work
- **J**ump to ISR address
- **R**un the ISR
- **R**esume original work

**5. Hardware vs Software Interrupt:**
- **H**ardware ‚Üí **I**nterrupt (H before I in alphabet)
- **S**oftware ‚Üí **S**ystem Call (Both start with S!)
