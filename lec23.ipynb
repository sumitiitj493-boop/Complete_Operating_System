---

**SRTF = Shortest Remaining Time First**

Ye basically **Preemptive SJF** ka fancy naam hai! ğŸ˜

**Funda:** Jis process ka **remaining CPU burst** sabse kam hai, usse CPU do!

### ğŸ”„ **Yaad Hai? #2: Preemption Ka Scene Kya Hai?**

---

**Preemptive = Beech mein CPU cheena ja sakta hai!** ğŸ”„

Jab bhi **naya process aaye**, check karo:

```json
IF (Naye process ka burst time) < (Current process ka REMAINING time)
    THEN: Current ko hatao, Naye ko CPU do! âš¡
ELSE:
    Current continue karega!
```
