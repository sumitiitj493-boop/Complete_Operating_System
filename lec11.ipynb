### üìå Flashback #1: Process aur PID Kya Tha?

**Process** = Program jo execute ho raha hai (Program in Execution)

**PID (Process ID)** = Har process ka **Unique Aadhar Number**! üÜî

| Concept | Meaning | Example |
|---------|---------|----------|
| Process | Running program | Chrome browser |
| PID | Unique identifier | PID = 1234 |

**Funda:** Jaise har insaan ka Aadhar unique hai, waise har process ka PID unique hai!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

### üìå Flashback #2: System Call Kya Hota Hai?

**System Call** = Process ka OS se **officially request** karna kisi kaam ke liye!

**Analogy:** Jaise tum bank mein jaake **form bharte ho** paise nikalne ke liye, waise process **system call** karta hai OS se kuch karwane ke liye!

**Important System Calls (aaj ke topic ke liye):**
- `fork()` / `create process` - Naya process banao
- `exit()` - Process khatam karo
- `wait()` - Parent apne children ka wait kare

## üìù Process Creation - Technical Definition

**Kya hota hai?**

Ek process **execution ke dauran** doosre naye processes create kar sakta hai using **create process system call**!

**Terminology:**
- **Parent Process** = Jo process naya process banata hai (Papa/Mummy)
- **Child Process** = Jo naya process bana (Baccha)
- **Process Tree** = Saare parent-child relationships ka structure (Family Tree)

## üßü‚Äç‚ôÇÔ∏è Zombie vs Orphan Process 

---

### üìå Formal Definitions

**Orphan Process:**  
A child process whose parent process has terminated before the child has finished its execution. Such child processes are adopted by `init/systemd`.

**Zombie Process:**  
A process that has completed execution but still has an entry left in the process table because its parent has not yet collected its exit status via `wait()`.

## üìù Final Takeaways:

### Process Creation:
1. **Parent creates Child** using `fork()` / create process system call
2. Forms a **Process Tree** structure
3. **PID** = Unique identifier for each process
4. **Execution:** Concurrent OR Parent waits
5. **Resources:** All OR Subset shared
6. **Address Space:** Duplicate OR New program

### Process Termination:
1. **Natural:** Process calls `exit()` after completing
2. **Forced:** Parent kills using `kill()` / terminate
3. **Status returned** to parent via `wait()`
4. **Resources deallocated** by OS
5. **Kill Reasons:** Resource exceeded, Task not needed, Cascading
