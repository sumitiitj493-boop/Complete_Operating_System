### ğŸ“š Flashback #1: Minimum Spanning Tree (MST) Kya Hota Hai?

Bhai, **Spanning Tree** matlab ek aisa tree jo:
- **Saare nodes ko connect kare** (koi bhi node akela na rahe)
- **N nodes ke liye exactly (N-1) edges ho** (bilkul utne hi wires, na zyada na kam)
- **Koi cycle na bane** (paisa waste nahi hona chahiye!)

Ab **MINIMUM Spanning Tree** matlab - in sabhi possible spanning trees mein se **sabse kam weight wala tree**! ğŸ†

## ğŸ’» Code Implementation - C++ Style!

Ab dekho bhai code mein kaise likhte hain:

```cpp
class Solution {
public:
    int spanningTree(int V, vector<vector<int>> adj[]) {
        
        // Step 1: Min-Heap banao (Priority Queue)
        // {weight, node} store karenge
        priority_queue<pair<int,int>, vector<pair<int,int>>, greater<pair<int,int>>> pq;
        
        // Step 2: Visited array banao
        vector<int> visited(V, 0);
        
        // Step 3: Starting point daalo (weight=0, node=0)
        pq.push({0, 0});
        
        // Step 4: Sum track karo
        int sum = 0;
        
        // Step 5: Jab tak PQ empty nahi, kaam karte raho!
        while (!pq.empty()) {
            // Sabse chhota weight wala nikalo
            auto it = pq.top();
            pq.pop();
            
            int node = it.second;   // Node number
            int wt = it.first;      // Edge weight
            
            // Agar already visited hai â†’ SKIP!
            if (visited[node]) continue;
            
            // Nahi toh, MST mein add karo!
            visited[node] = 1;      // Mark visited
            sum += wt;              // Weight add karo
            
            // Saare unvisited neighbors ko PQ mein daalo
            for (auto& edge : adj[node]) {
                int adjNode = edge[0];      // Adjacent node
                int edgeWt = edge[1];       // Edge weight
                
                if (!visited[adjNode]) {
                    pq.push({edgeWt, adjNode});
                }
            }
        }
        
        return sum;  // Final MST weight!
    }
};
```

<p style="color: #c9d1d9;">

2. **Visited Array:** Kaun MST mein hai, track karne ke liye<br><br>
3. **pq.push({0, 0}):** Node 0 se shuru, with weight 0<br><br>
4. **if (visited[node]) continue:** Already visited? SKIP!<br><br>
5. **sum += wt:** MST mein add ho gaya, weight bhi add karo!<br><br>
6. **for loop:** Saare neighbors check karo, unvisited walon ko PQ mein daalo
