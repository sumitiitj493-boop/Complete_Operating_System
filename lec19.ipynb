## <span style="color: #58a6ff;">Technical Breakdown - CPU Burst Kya Hai?</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #58a6ff;">

### <span style="color: #58a6ff;">CPU BURST</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Definition:** Wo time jab process CPU use kar raha hai apni execution ke liye.

**Simple Words Mein:** Process ka "kaam karne ka time" - jab wo actually CPU pe calculations kar raha hai.

**Examples of CPU Burst Operations:**
- `LOAD` - Memory se data load karna
- `STORE` - Memory mein data store karna  
- `ADD` - Addition karna
- `INCREMENT` - Value badhana
- Calculations, logic operations, etc.

</p>

</div>

<br>

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #a371f7;">

### <span style="color: #a371f7;">I/O BURST</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Definition:** Wo time jab process kisi I/O operation ke complete hone ka wait kar raha hai.

**Simple Words Mein:** Process ka "waiting time" - jab wo CPU use NAHI kar raha, kisi external cheez ka wait kar raha hai.

**Examples of I/O Burst:**
- `READ FROM FILE` - File se data padhna
- `WRITE TO FILE` - File mein likhna
- User input ka wait
- Network response ka wait
- Printer se print hone ka wait

</p>

</div>

</div>

## <span style="color: #58a6ff;">Technical Breakdown</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #58a6ff;">

### <span style="color: #58a6ff;">CPU SCHEDULER (Short-Term Scheduler)</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Kaam:** Jab CPU idle ho jaaye, Ready Queue mein se ek process SELECT karna.

**Simple Words:** Ye wo banda hai jo decide karta hai - "Bhai, ab tera number hai CPU use karne ka!"

```json
Ready Queue: [P1, P2, P3, P4, P5]
                    ^
                    |
        CPU Scheduler: "P3, tu ja!"
```

</p>

</div>

<br>

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #3fb950;">

### <span style="color: #3fb950;">DISPATCHER</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Kaam:** Selected process ko ACTUALLY CPU ka control dena.

**Dispatcher kya karta hai:**
1. Context Switch karta hai (purani process ka state save, nayi ka load)
2. User mode mein switch karta hai
3. Process ko sahi jagah pe jump kara deta hai continue karne ke liye

**Zaruri:** Dispatcher **bahut FAST** hona chahiye! Kyunki har baar process switch pe ye run hota hai.

</p>

</div>

<br>

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #f0883e;">

### <span style="color: #f0883e;">DISPATCH LATENCY</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Definition:** Wo time jo dispatcher leta hai ek process ko rokne mein aur doosri ko start karne mein.

**Formula:** Dispatch Latency = Time(Stop Process A) + Time(Start Process B)

**Goal:** Dispatch Latency **minimum** honi chahiye for efficient system!

Jaise shaadi mein agar waiter slow hai, toh sab mehmaan irritate! Same here - slow dispatcher = slow system.

## <span style="color: #58a6ff;">Dono Kab Use Hote Hain?</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #58a6ff;">

### <span style="color: #58a6ff;">Non-Preemptive Kab Acha?</span>

<p style="color: #c9d1d9;">

- Simple systems mein
- Jahan process zyada time na le
- Batch processing systems
- Jahan response time critical na ho

</p>

</div>

<br>

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #3fb950;">

### <span style="color: #3fb950;">Preemptive Kab Acha?</span>

<p style="color: #c9d1d9;">

- Interactive systems (jaise tumhara phone, laptop)
- Real-time systems (jahan timing critical ho)
- Multi-user systems
- Jahan high-priority tasks ko jaldi execute karna ho

</p>

</div>

<br>

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #f0883e;">

### <span style="color: #f0883e;">Important Note</span>

<p style="color: #c9d1d9;">

**Bhai, ye dono scheduling TYPES hain, ALGORITHMS nahi!**

- FCFS, SJF (Non-Preemptive) - Ye algorithms hain
- Round Robin, SRTF (Preemptive) - Ye bhi algorithms hain

Preemptive/Non-Preemptive batata hai ki **CPU cheeni ja sakti hai ya nahi**, algorithm batata hai **kis process ko CPU milegi**.

</p>
