### <span style="color: #58a6ff;">1. CPU Burst Time Kya Hai?</span>

<p style="color: #c9d1d9;">

**Burst Time = Wo time jitna process ko CPU chahiye apna kaam khatam karne ke liye.**

Jaise canteen mein ek plate serve karne mein 5 minute lagte hain - wo hai burst time!

```
Process P1: Burst Time = 24ms (Bada kaam hai, time lagega)
Process P2: Burst Time = 3ms  (Chhota kaam hai, jaldi hoga)
```

</p>

---

### <span style="color: #58a6ff;">2. Ready Queue Kya Hai?</span>

<p style="color: #c9d1d9;">

**Ready Queue = Wo line jismein saare processes khade hain, CPU ka wait kar rahe hain!**

Jaise bank mein log line mein khade hote hain counter ke liye - wahi Ready Queue hai!

</p>

---

### <span style="color: #58a6ff;">3. Arrival Time Kya Hai?</span>

<p style="color: #c9d1d9;">

**Arrival Time = Process kab Ready Queue mein aaya?**

Jaise tumne 10:00 baje bank mein entry li - wo tumhara arrival time!

</p>

---

### <span style="color: #58a6ff;">4. FIFO Queue Kya Hai?</span>

<p style="color: #c9d1d9;">

**FIFO = First In, First Out**

Jo pehle aayega, wo pehle jaayega! Simple!

## <span style="color: #58a6ff;">1. CPU UTILIZATION</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #58a6ff;">

### <span style="color: #58a6ff;">Definition:</span>

<p style="color: #c9d1d9; font-size: 16px;">

**CPU Utilization = CPU kitna busy hai? Kitne percent time kaam kar raha hai?**

</p>

### <span style="color: #3fb950;">Real Life Example:</span>

<p style="color: #c9d1d9;">

Socho ek **Chef** hai kitchen mein:
- Agar Chef 8 ghante mein se **6 ghante khana bana raha** hai = **75% utilization**
- Agar Chef **2 ghante baitha** raha (idle) = **Time waste!**

**Goal:** CPU ko maximum busy rakhna! Idle CPU = Paisa barbaad!

</p>

### <span style="color: #f0883e;">Technical Points:</span>

<p style="color: #c9d1d9;">

- **Conceptually:** 0% to 100% range
- **Real System:** 40% (light load) to 90% (heavy load)
- **Goal:** Keep CPU as BUSY as possible!
- **Higher is better!** â¬†ï¸

## <span style="color: #3fb950;">2. THROUGHPUT</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #3fb950;">

### <span style="color: #3fb950;">Definition:</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Throughput = Per unit time kitne processes COMPLETE hue?**

</p>

### <span style="color: #58a6ff;">Real Life Example:</span>

<p style="color: #c9d1d9;">

Socho ek **Factory** hai jo phones banati hai:
- **1 ghante mein 100 phones** = Throughput = 100 phones/hour
- **Zyada throughput = Zyada production = Better!**

CPU ke case mein:
- **1 second mein 50 processes complete** = Throughput = 50 processes/sec

</p>

### <span style="color: #f0883e;">Technical Points:</span>

<p style="color: #c9d1d9;">

- **Formula:** Throughput = Number of processes completed / Time
- **Unit:** Processes per unit time (per second, per hour, etc.)
- **Higher is better!** â¬†ï¸
- Measures **PRODUCTIVITY** of CPU

</p>

## <span style="color: #f0883e;">3. TURNAROUND TIME (TAT)</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #f0883e;">

### <span style="color: #f0883e;">Definition:</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Turnaround Time = Process submit karne se leke complete hone tak ka TOTAL time**

</p>

### <span style="color: #58a6ff;">Real Life Example:</span>

<p style="color: #c9d1d9;">

Socho tum **Zomato pe order karte ho**:
- Order diya: **7:00 PM**
- Khana aaya: **7:45 PM**
- **Turnaround Time = 45 minutes**

Isme sab include hai:
- Restaurant ne order accept kiya (waiting)
- Khana bana (execution)
- Delivery boy travel kiya (more waiting)

</p>

### <span style="color: #3fb950;">Formula:</span>

```
Turnaround Time = Completion Time - Arrival Time

TAT = CT - AT
```

### <span style="color: #a371f7;">Components:</span>

<p style="color: #c9d1d9;">

TAT includes:
1. â³ Waiting to get into memory
2. â³ Waiting in Ready Queue
3. âš¡ Executing on CPU
4. â³ Doing I/O operations

**Lower is better!** â¬‡ï¸ (Customer jaldi khush!)

</p>

## <span style="color: #a371f7;">4. WAITING TIME (WT)</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #a371f7;">

### <span style="color: #a371f7;">Definition:</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Waiting Time = Ready Queue mein wait karne ka TOTAL time**

</p>

### <span style="color: #58a6ff;">Real Life Example:</span>

<p style="color: #c9d1d9;">

Socho tum **Bank ki line mein khade ho**:
- Line mein lage: **10:00 AM**
- Counter pe pahunche: **10:30 AM**
- **Waiting Time = 30 minutes**

Note: Counter pe kaam karne ka time (5 min) waiting time mein **NAHI** aata!

</p>

### <span style="color: #3fb950;">Important Point:</span>

<p style="color: #c9d1d9;">

**Scheduling algorithm SIRF waiting time ko affect karta hai!**

Kyun? Because:
- Process ka execution time (burst time) = Process pe depend karta hai
- I/O time = Process pe depend karta hai
- **Scheduler sirf decide karta hai ki CPU kisko de!**

Toh scheduler ki efficiency check karne ke liye **Waiting Time** best metric hai!

</p>

### <span style="color: #f0883e;">Formula:</span>

```
Waiting Time = Turnaround Time - Burst Time

WT = TAT - BT
```

<p style="color: #c9d1d9;">

**Lower is better!** â¬‡ï¸ (Kam wait = Happy process!)

</p>

## <span style="color: #f85149;">5. RESPONSE TIME (RT)</span>

---

<div style="background-color: #161b22; padding: 20px; border-radius: 10px; border-left: 5px solid #f85149;">

### <span style="color: #f85149;">Definition:</span>

<p style="color: #c9d1d9; font-size: 16px;">

**Response Time = Request submit karne se FIRST RESPONSE milne tak ka time**

**Note:** Ye complete hone ka time NAHI hai, sirf pehla response!

</p>

### <span style="color: #58a6ff;">Real Life Example:</span>

<p style="color: #c9d1d9;">

Socho tum **Video file pe double-click** karte ho:
- Click kiya: **10:00:00**
- Video player khula (loading screen dikhaya): **10:00:02** (2 seconds)
- Video fully load hua: **10:00:10** (10 seconds)

**Response Time = 2 seconds** (pehla response - player khulna)
**Turnaround Time = 10 seconds** (poora video load hona)

Difference samjhe? Response time sirf **FIRST reaction** measure karta hai!

</p>

### <span style="color: #3fb950;">Formula:</span>

```
Response Time = First Response Time - Arrival Time

RT = Time of First Response - AT
```

### <span style="color: #f0883e;">When is RT important?</span>

<p style="color: #c9d1d9;">

- **Interactive Systems** mein (jaise games, UI applications)
- User ko **instant feedback** chahiye
- Turnaround time device speed pe depend karta hai (slow screen = slow output)
- Response time system ki **responsiveness** batata hai!

**Lower is better!** â¬‡ï¸

</p>
