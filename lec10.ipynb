### üìå Flashback #1: Multiprogramming Kya Tha?

**Multiprogramming** = Multiple programs ko memory mein rakho, taaki CPU kabhi **idle na baithe**!

**Objective:** CPU utilization **maximize** karna! 

**Analogy:** Tum ek **Zomato delivery boy** ho. Agar ek order deliver karke ghar baith jaoge, to paise kaise banenge? 
Isliye tum ek saath **multiple orders** uthate ho - ek deliver karte waqt doosra ready rehta hai!

**Funda:** CPU ko hamesha kuch na kuch kaam mein busy rakhna hai!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

### üìå Flashback #2: Time Sharing Kya Tha?

**Time Sharing** = CPU ka time **chhote-chhote slices** mein divide karo aur har process ko thoda-thoda time do!

**Objective:** User ko lagni chahiye ki **saare programs ek saath chal rahe hain**!

**Analogy:** Tum **3 WhatsApp chats** pe reply kar rahe ho:
- GF ko 1 line likhi ‚úçÔ∏è
- Dost ko 1 line likhi ‚úçÔ∏è  
- Mom ko 1 line likhi ‚úçÔ∏è
- Phir GF ko wapas...

Teeno ko lagta hai tum unse hi baat kar rahe ho, but tum toh **rapidly switch** kar rahe ho! üòÑ

**Funda:** Itna fast switch karo ki user ko pata hi na chale!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## üìù Process Scheduler - Kya Hai Ye Cheez?

**Process Scheduler** = OS ka wo component jo **decide karta hai** ki:

1. Kaunsa process **pehle run** karega?
2. Kitni **der tak** run karega?
3. Kab **switch** karna hai doosre process pe?

**Basically:** CPU ke time ka **traffic police**! üö¶

</div>

## 1Ô∏è‚É£ Job Queue (Long-term Queue)

<div style="background-color: #161b22; border-left: 5px solid #a371f7; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** System mein jitne bhi processes hain - **SAB** yahan hain!

</div>

**Kya hai?**
- Jab koi naya process create hota hai, pehle **Job Queue** mein jaata hai
- Ye **waiting list** hai - abhi execute nahi ho rahe, par system mein hain
- Yahan se processes select hoke **Ready Queue** mein jaate hain

**Analogy:** Movie theater ke bahar **lambi line** - sabko andar jaana hai, par pehle ticket check hoga!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## 2Ô∏è‚É£ Ready Queue (Short-term Queue)

<div style="background-color: #161b22; border-left: 5px solid #3fb950; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Main memory mein hain, **ready** hain, bas **CPU ka wait** kar rahe hain!

</div>

**Kya hai?**
- Processes jo **execute hone ke liye taiyaar** hain
- Memory mein loaded hain
- Bas CPU assign hone ka wait kar rahe hain
- Jaise hi CPU free hua, **Scheduler** inme se ek ko pick karega!

**Analogy:** Restaurant mein **kitchen ke paas order tickets** - chef bas ek complete kare aur next uthaye!

## üìù Context Switch - Technical Definition

**Context Switch** = CPU ko ek process se doosre process pe switch karna, jisme:

1. **State Save** - Current process ka context (PCB) save karo
2. **State Restore** - New process ka context (PCB) restore karo

**Context kya hota hai?**
- Program Counter (next instruction address)
- CPU Registers ki values
- Process State
- Memory management info
- Basically... **PCB mein jo bhi hai!**

## üö® WARNING: Context Switch Time is OVERHEAD!

**Overhead** = Extra cost/time jo kisi kaam mein lagti hai, but actual kaam nahi hai!

**Problem:**
- Jab context switch ho raha hai, **CPU koi useful kaam NAHI kar raha!**
- Sirf save/restore ho raha hai
- Koi process execute nahi ho raha

**Yaad karo:** Multiprogramming ka goal tha - **CPU hamesha busy rahe**
But context switch ke time - **CPU sirf switching mein busy hai, actual kaam mein nahi!**

## üìä Context Switch Time - Factors

**Kitna time lagta hai?** Depend karta hai:

| Factor | Impact |
|--------|--------|
| **Memory Speed** | Fast memory = Fast switch |
| **Number of Registers** | Zyada registers = Zyada save/restore time |
| **Special Instructions** | Single instruction for all registers = Fast |
| **OS Complexity** | Simple OS = Fast switch |

**Typical Speed:** Few **milliseconds** (bahut fast!)
