### **Pehle Ek Kahani Suno...**

Socho tumhari **Mummy** ne bola: "Ja, kitchen mein **haldi dhundh ke la**!"

## **Technical Definition:**

**Thread Cancellation** = Ek thread ko **uska kaam complete hone se pehle** terminate kar dena.

Simple Bhasha Mein: Thread ki **naukri khatam** karna beech mein hi!

**Target Thread:** Jis thread ko cancel karna hai, usse **Target Thread** bolte hai. Ye basically **victim** hai jo cancel hone wala hai!

## **Type 1: Asynchronous Cancellation** âš¡

### **The "Boss Ne Seedha Nikal Diya" Approach**

**Kya hota hai?**
- Ek thread **immediately terminate** kar deta hai target thread ko
- **Koi warning nahi**, koi notice nahi, seedha OUT!
- Target thread ke paas **koi choice nahi** - ye victim hai!

**Real-Life Analogy:**

Socho **Office Scene**:
- Employee (Thread) apna kaam kar raha hai
- Boss aaya: "Tu fired! Abhi ke abhi nikal!"
- Employee ke haath mein **half-complete file** thi
- Company ka **laptop uske paas** tha
- Pending **bills uske paas** the

**Result:** Employee chala gaya, lekin **sab cheezein adhoori** reh gayi!

## **Type 2: Deferred Cancellation** ğŸ¤

### **The "Notice Period Wala" Approach**

**Kya hota hai?**
- Target thread **periodically check** karta hai: "Kya mujhe cancel hona chahiye?"
- Agar haan, toh **apne aap ko orderly fashion** mein terminate karta hai
- Thread ke paas **power hai** - wo decide karega KAISE aur KAB cancel hona hai

**Real-Life Analogy:**

**Proper Resignation Scene:**
- HR ne bola: "Bhai, company tumhe release kar rahi hai, 1 month notice period hai"
- Employee ne apni **pending files complete ki**
- Company ka **laptop return kiya**
- **Handover** de diya next person ko
- Phir **gracefully exit** kar gaya

**Result:** Sab kuch **proper tarike se** band hua!

# **WRAP-UP: Moral of the Story** ğŸ“

---

## **Thread Cancellation:**

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; margin: 10px 0;">

**"Kisi ko beech mein nikalna hai toh GRACEFULLY nikalo, warna resources aur data ka satyanash ho jaayega!"**

- **Asynchronous** = Seedha fire, risky hai
- **Deferred** = Notice period, safe hai
- **Always prefer Deferred Cancellation!**

</div>

---

## **CPU Scheduling:**

<div style="background-color: #0d1117; padding: 15px; border-radius: 10px; margin: 10px 0;">

**"CPU ko kabhi free mat baithne do - jab ek wait kare, dusre ko de do!"**

- CPU = Bathroom, Processes = Family members
- Problem: I/O wait mein CPU idle rehta hai
- Solution: Switch CPU to another process!
- Different algorithms decide kaun pehle jaayega

</div>
