### ğŸ“Œ **Concept 1: FCFS Scheduling Algorithm Kya Hai?**

**Analogy:** Soch bhai, **Railway Ticket Counter** pe khade ho. Jo pehle aaya, uska pehla number. Simple rule - no VIP treatment, no shortcuts!

**Technical Definition:**
- Jo process **pehle Ready Queue mein aata hai**, usse **pehle CPU milta hai**
- **Non-Preemptive** hai - matlab ek baar CPU mil gaya, toh poora kaam hone tak koi nahi cheenay sakata
- Sabse **simple** scheduling algorithm hai

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; border: 1px solid #30363d; color: #c9d1d9;">

### ğŸ“Œ **Concept 2: Important Terms - Ye Samajhna Zaroori Hai!**

| Term | Desi Explanation | Example |
|------|------------------|----------|
| **Arrival Time (AT)** | Process kab aaya Ready Queue mein? | Student canteen mein 12:00 baje aaya |
| **Burst Time (BT)** | CPU ko kitna time chahiye execution ke liye? | Order complete karne mein 5 min lagenge |
| **Completion Time (CT)** | Process kab finish hua completely? | Student 12:15 baje nikla canteen se |
| **Turnaround Time (TAT)** | Total time from arrival to completion | 12:15 - 12:00 = 15 min total laga |
| **Waiting Time (WT)** | Kitna time sirf wait kiya CPU ke liye? | 15 min - 5 min (actual order time) = 10 min wait |

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; border: 1px solid #30363d; color: #c9d1d9;">

### ğŸ“Œ **Concept 3: Formulas - Yaad Karo Nahi, Samjho!**

**ğŸ¯ Turnaround Time (TAT):**
```json
TAT = Completion Time - Arrival Time
```
**Matlab:** Jab se aaye the tab se jab tak nikle - poora time!

---

**ğŸ¯ Waiting Time (WT):**
```json
WT = Turnaround Time - Burst Time
```
**Matlab:** Poore time mein se actual kaam ka time nikaal do - jo bacha wo wait time!

---

**ğŸ¯ Efficiency:**
```json
Efficiency = (Useful Time / Total Time) Ã— 100%
```
