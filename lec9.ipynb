### üìå Flashback #1: Program vs Process - Kya Farak Hai Bhai?

| Concept | Kya Hai? | Real Life Example |
|---------|----------|-------------------|
| **Program** | Code jo hard disk pe pada hai, **passive** hai, kuch nahi kar raha | Recipe book jo shelf pe padi hai üìñ |
| **Process** | Jab program **execute** ho raha hai, **active** hai, kaam ho raha hai | Jab actually khaana ban raha hai üç≥ |

**Simple Funda:** 
> *"Program in Execution = Process"*  
> Matlab jab code run hota hai, tab wo process ban jaata hai!

### üìå Flashback #2: CPU/Processor Kya Hai?

**CPU (Central Processing Unit)** = Computer ka **Dimaag** üß†

- Ye actually instructions execute karta hai
- Ek time pe primarily ek hi kaam kar sakta hai (single core mein)
- Process ko execute karne ke liye CPU chahiye hi chahiye!

**Soch lo:** CPU ek **chef** hai kitchen mein. Kitne bhi orders aa jaayein, ek chef ek time pe ek hi dish bana sakta hai!

## üîµ State 1: NEW State

<div style="background-color: #161b22; border-left: 5px solid #58a6ff; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Process is being created / Process ban raha hai

</div>

**Kya ho raha hai?**
- Tumne program run kiya (double-click ya command)
- OS bol raha hai: *"Accha, naya process banana hai? Ruk, setup karta hoon..."*
- Memory allocate ho rahi hai, PCB ban raha hai

**Real Life:** Zomato pe order place kiya, abhi restaurant tak notification nahi gayi!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## üü¢ State 2: READY State

<div style="background-color: #161b22; border-left: 5px solid #3fb950; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Process is waiting to be assigned to a processor / CPU ka wait kar raha hai

</div>

**Kya ho raha hai?**
- Process ban chuka hai (NEW complete)
- Ab **line mein khada hai**, CPU free hone ka wait kar raha hai
- Jaise hi CPU milega, execute hona shuru!

**Real Life:** Restaurant ne order accept kar liya, par chef busy hai doosre orders mein. Tumhara order **queue mein** hai!

**Key Point:** Process **READY hai but RUNNING nahi** - matlab taiyaar hai par CPU nahi mila abhi!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## üü† State 3: RUNNING State

<div style="background-color: #161b22; border-left: 5px solid #f0883e; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Instructions are being executed / Instructions execute ho rahe hain

</div>

**Kya ho raha hai?**
- **CPU mil gaya!** üéâ
- Process ke instructions line by line execute ho rahe hain
- Actual kaam ho raha hai!

**Real Life:** Chef ne tumhara Biryani banana **shuru kar diya**! Rice boil ho rahi, masale dal rahe, sab kuch ho raha hai!

**Important:** Ek CPU core pe ek time mein **sirf EK hi process** RUNNING state mein ho sakta hai!

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## üü£ State 4: WAITING State (Blocked State bhi bolte hain)

<div style="background-color: #161b22; border-left: 5px solid #a371f7; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Process is waiting for some event to occur (I/O completion, signal, etc.)

</div>

**Kya ho raha hai?**
- Process ko kuch **bahar se chahiye** (file read, user input, network data)
- CPU ke paas kuch nahi karna, toh **CPU chhod deta hai**
- Jab tak required cheez nahi milti, yahan wait karega

**Real Life:** Chef biryani bana raha tha, suddenly pata chala **Basmati Rice khatam hai!** üò±  
Ab order WAIT pe hai jab tak rice nahi aa jaata!

**Events ka wait:**
- I/O operation complete hona (file read/write)
- User se input aana
- Network se data aana
- Koi signal receive hona

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

## üî¥ State 5: TERMINATED State (Exit State)

<div style="background-color: #161b22; border-left: 5px solid #f85149; padding: 15px; color: #c9d1d9; margin: 10px 0;">

**Definition:** Process has finished execution / Process khatam ho gaya

</div>

**Kya ho raha hai?**
- Process ne apna **saara kaam kar liya**
- Ya phir process ko **forcefully terminate** kar diya (kill command, error, etc.)
- Ab resources free honge, memory wapas milegi

**Real Life:** Biryani deliver ho gayi, tumne kha li, order **complete!** üéâ  
Ab wo order system se hatt jayega!

## üîÄ Transitions Explained - Kaun Kahan Jaata Hai?

### 1Ô∏è‚É£ NEW ‚û°Ô∏è READY (Admitted)
- Process create ho gaya
- OS ne admit kar liya
- Ab ready queue mein wait karega

### 2Ô∏è‚É£ READY ‚û°Ô∏è RUNNING (Scheduler Dispatch)
- **Scheduler** decide karta hai ki kaun chalega
- CPU allocate hota hai
- Execution shuru!

### 3Ô∏è‚É£ RUNNING ‚û°Ô∏è TERMINATED (Exit)
- Process ne apna kaam **successfully complete** kar liya
- Ya error aa gaya, ya kill kar diya
- Game over! üéÆ

### 4Ô∏è‚É£ RUNNING ‚û°Ô∏è READY (Interrupt)
- **Higher priority process** aa gaya! üö®
- Ya **time quantum khatam** ho gaya (time sharing mein)
- Current process ko side mein rakho, baad mein continue karenge

### 5Ô∏è‚É£ RUNNING ‚û°Ô∏è WAITING (I/O or Event Wait)
- Process ko kuch **bahar se chahiye**
- I/O request ki (file read, user input)
- CPU chhod do, wait karo jab tak nahi milta

### 6Ô∏è‚É£ WAITING ‚û°Ô∏è READY (I/O or Event Completion)
- Jo wait kar rahe the wo **mil gaya!**
- Ab ready ho gaye phir se run karne ke liye
- Ready queue mein wapas jao, CPU ka wait karo

## üì¶ PCB Components - Ek Ek Karke Samjho!

### 1Ô∏è‚É£ Process ID (PID)

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Har process ka **unique number** - uski pehchaan!

**Analogy:** Tumhara Aadhar Number - poore India mein unique!

**Use:** OS isi number se process ko identify karta hai

</div>

### 2Ô∏è‚É£ Process State

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Process abhi kis state mein hai (New/Ready/Running/Waiting/Terminated)

**Analogy:** Zomato order status - Preparing/On the way/Delivered

**Use:** OS ko pata rehta hai ki process ka kya scene hai

</div>

### 3Ô∏è‚É£ Program Counter (PC)

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** **Next instruction ka address** jo execute hogi

**Analogy:** Book mein bookmark - kahan tak padh liya, aage kahan se shuru karna hai

**Use:** Jab process wapas run hoga, pata rehta hai kahan se continue karna hai!

**Example:** Line 1, 2, 3 execute ho gaye, PC = Line 4 ka address

</div>

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

### 4Ô∏è‚É£ CPU Registers

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Process ke **register values** save rehte hain

**Types:** Index registers, Stack pointers, General purpose registers, etc.

**Analogy:** Exam mein tum rough work kar rahe the, suddenly break lena pada - rough sheet save karoge na wapas continue karne ke liye!

**Use:** Context switch ke time register values save hote hain, taaki baad mein restore kar sakein

</div>

### 5Ô∏è‚É£ CPU Scheduling Information

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** 
- Process ki **priority**
- Scheduling queue ke **pointers**
- Other scheduling parameters

**Analogy:** Hospital mein kuch patients **emergency** hain (high priority), kuch **OPD** (normal priority)

**Use:** OS decide karta hai ki kaun pehle run karega based on priority!

</div>

### 6Ô∏è‚É£ Memory Management Information

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Process ko kitni aur kahan **memory mili** hai

**Contains:**
- Base and limit registers
- Page tables / Segment tables
- Memory limits

**Analogy:** Hostel mein tumhe kaunsa room mila, kitni jagah hai - sab info!

</div>

</div>

<div style="background-color: #0d1117; padding: 20px; border-radius: 10px; color: #c9d1d9;">

### 7Ô∏è‚É£ Accounting Information

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Process ne kitne **resources use kiye** - hisaab kitaab!

**Tracks:**
- CPU time used
- Real time elapsed
- Time limits
- Job/Process numbers

**Analogy:** Mobile ka data usage tracker - kitna data use hua, kitna bacha!

**Use:** Billing, performance analysis, fair resource allocation

</div>

### 8Ô∏è‚É£ I/O Status Information

<div style="background-color: #161b22; padding: 15px; border-radius: 8px; margin: 10px 0;">

**Kya hai:** Process ko kaunse **I/O devices assigned** hain, kaunsi **files open** hain

**Contains:**
- List of open files
- I/O devices allocated

**Analogy:** Office mein tumhe kaunsi printer assigned hai, kaunsa phone extension mila hai!

**Use:** Resource management, ensuring proper I/O handling

</div>
