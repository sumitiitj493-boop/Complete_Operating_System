## 1ï¸âƒ£ Zero Capacity Buffer - "Direct Handover Only!"

**Kya hai:**
- Queue ka size = **0**
- **Koi message wait nahi** kar sakta
- Sirf ek message at a time pass ho sakta hai

**Rule:** Sender **MUST block** until receiver receives!

**Kyun?** 
- Agar sender blocked nahi hoga, second message bhejega
- But buffer mein jagah hi nahi hai!
- Toh pehla message receive hona **mandatory** hai!

**Analogy:** Class mein **chit pass** karna - directly haath se haath mein dena padega, table pe rakh ke nahi chhodna! ğŸ“

## 2ï¸âƒ£ Bounded Capacity Buffer - "Limited Seats Available!"

**Kya hai:**
- Queue ka size = **n** (fixed number)

**Rules:**
- Buffer **not full** â†’ Sender sends, continues working (no blocking)
- Buffer **FULL** â†’ Sender **MUST block** until space available!

**Analogy:** 
- **Rajdhani Express** with 50 seats
- Seats available â†’ Book karo, niklo! ğŸš‚

## 3ï¸âƒ£ Unbounded Capacity Buffer - "Infinite Space!"

## ğŸ“ Final Takeaways:

### Synchronization (Blocking vs Non-Blocking):
1. **Blocking Send:** Wait until message received
2. **Non-Blocking Send:** Send and continue
3. **Blocking Receive:** Wait until message available
4. **Non-Blocking Receive:** Get message OR null immediately

### Buffering:
1. **Zero Capacity:** No waiting, direct transfer, sender always blocks
2. **Bounded Capacity:** n messages can wait, sender blocks when full
3. **Unbounded Capacity:** Infinite waiting, sender never blocks

### Sockets:
1. Socket = **IP Address : Port Number**
2. Ports **< 1024** = Reserved (well-known)
3. Ports **> 1024** = For client processes
4. Server **listens**, Client **connects**
5. Used specifically for **Client-Server** systems
