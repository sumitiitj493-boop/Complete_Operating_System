## üìù RPC - Technical Definition

**Remote Procedure Call (RPC):**
> A **protocol** that allows a program to request a **service** from another program located on a **different computer** in a network, **without understanding network details**!

**Key Points:**
1. **Remote** = Doosre system pe (not local)
2. **Procedure** = Function/Service
3. **Call** = Request karna
4. **Network details hidden** = Developer ko network ki chinta nahi!

## üßë‚Äçüíº Stub Kya Hai? - The Delivery Boys!

**Stub** = Client aur Server ke beech ka **middleman**!

### Client Stub (Tumhari Side Ka Delivery Boy)

**Kya karta hai:**
1. **Locates server port** - Restaurant ka address dhundho
2. **Marshals parameters** - Order properly pack karo
3. **Sends message** - Bhej do network pe!

---

### Server Stub (Unki Side Ka Delivery Boy)

**Kya karta hai:**
1. **Receives message** - Order receive karo
2. **Unmarshals parameters** - Package kholo, order dekho
3. **Invokes procedure** - Chef ko batao, banana shuru!
4. **Returns result** - Pizza wapas bhejo

```json
1. Client sends request
2. Server executes procedure
3. Server sends ACK (acknowledgment)
4. Client receives ACK ‚Üí Done!
   
   If NO ACK received:
   Client sends request AGAIN until ACK milta hai!
```

## üìù Final Takeaways:

### What is RPC:
1. Protocol for **remote communication** between processes
2. Processes on **different systems** connected via **network**
3. Uses **message-based** communication (not shared memory)
4. Hides **network details** from developer

### Key Components:
1. **Client Stub** - Locates server, marshals params, sends message
2. **Server Stub** - Receives message, unmarshals, invokes procedure
3. **Matchmaker** - Directory service for port lookup
4. **RPC Daemon** - Listens on ports, handles requests

### Three Issues:
1. **Data Representation** ‚Üí Solved by **XDR** (universal format)
2. **Network Errors** ‚Üí Solved by **Exactly Once** (ACK system)
3. **Port Binding** ‚Üí Solved by **Matchmaker** (dynamic lookup)
